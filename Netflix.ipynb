```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('netflix_titles.csv')
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.fillna('UNKNOWN',inplace=True)
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data
```

```python
data['date_added'] = pd.to_datetime(data['date_added'].str.strip(),errors='coerce')
```

```python
data
```

```python

# Extract numeric duration for movies only (safe version)
data['duration_mins'] = data['duration'].str.extract(r'(\d+)').astype(float)


# Now calculate IQR on 'duration_mins'
Q1 = data['duration_mins'].quantile(0.25)
Q3 = data['duration_mins'].quantile(0.75)
IQR = Q3 - Q1

data = data[(data['duration_mins'] >= (Q1 - 1.5 * IQR)) & (data['duration_mins'] <= (Q3 + 1.5 * IQR))]
```

```python
data
```

```python
from collections import Counter
import itertools

# Explode genre list and count
data['listed_in'] = data['listed_in'].fillna('').str.split(', ')
genre_counts = Counter(itertools.chain.from_iterable(data['listed_in']))
genre_counts.most_common(10)
```

```python
director = data.copy()
director['director'] = director['director'].fillna('Unknown')
top_directors = director[director['director'] != 'Unknown']['director'].value_counts().head(10)
top_directors = top_directors.reset_index()
top_directors.columns = ['director', 'number_of_titles']

top_directors
```

```python
label = ['Movie', 'TV show']
size = data['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [ 0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=label, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```

```python
top_countries = data['country'].value_counts().head(10)
colors = [(1 - i / len(top_countries), 0, 0) for i in range(len(top_countries))]
fig, ax = plt.subplots()
bars = ax.bar(top_countries.index, top_countries.values, color=colors)
plt.title('Top 10 Countries by Number of Netflix Titles')
plt.ylabel('Number of Titles')
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np

data['year_added'] = data['date_added'].dt.year
titles_per_year = data['year_added'].value_counts().sort_index()
colors = plt.cm.viridis(np.linspace(0, 1, len(titles_per_year)))

plt.plot(titles_per_year.index, titles_per_year.values, marker='o', color='black')  
for i, year in enumerate(titles_per_year.index):
    plt.scatter(year, titles_per_year[year], color=colors[i], zorder=5)

plt.title('Netflix Titles Added per Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()
```

```python
sns.countplot(y='rating', data=data, order=data['rating'].value_counts().index)
plt.title('Content Ratings Distribution')
plt.show()
```
